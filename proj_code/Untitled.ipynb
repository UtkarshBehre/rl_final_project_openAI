# This document gives some details on how to set up account and use CCR and help with some common things
​
## Setup account in CCR
1. To access CCR you need to be in the UB network or connected over VPN. You can get information on how to and which VPN clients are supported [here](http://www.buffalo.edu/ubit/service-guides/connecting/vpn/computer.html). It also shows how to set up and connect to UB network.
2. Once you are done with that you can start setting up your CCR account by following the steps [here](https://ubccr.freshdesk.com/en/support/solutions/articles/13000054633-creating-a-ccr-account).
3. Once done login to your account and setup 2 factor authentication using google authenticator or any other application which is listed in 2 factor authentication space. Once you have set this up, for every consequent login you have to type in your password followed by the OTP in your application to login to your account.
4. You will have to create a ssh key pair for which you can refer [here](https://docs.rightscale.com/faq/How_Do_I_Generate_My_Own_SSH_Key_Pair.html).
5. After you are done creating your account you have to let the professor who is allocating the resources to you know so that they can add you under the project and give you access to CCR.
6. You will have to wait for a day or so after the professor has added you to the project for you to be able to access any resources.
7. You can see your project and all it's users, publications under the project and all information [here](https://coldfront.ccr.buffalo.edu/). This is where you will be able to see you are part of a project after your professor adds you.
8. You can queue and ask for a session by going to on Demand(the interactive and easy web portal which gives access to CCR's HPC resource) by going [here](https://ondemand.ccr.buffalo.edu/)
9. Click on interactive apps on the top header. You should be able to see jupyter 3.6 in the options. You can click on that. You then input fields for **slurm** account as *cse510* in our case or you can leave it empty. **partition** as *GPU* and **QOS** as also *GPU*. Rest of the things you fill out based on your need.
​
You can read FAQs and other people's threads and forums [here](https://ubccr.freshdesk.com/support/home).
​
## Advanced user guide
The further commands go with assumptions that user knows basic unix commands and understands what user is doing. It is very important to understand the commands being run as these commands will be run on a cluster being shared with many different people and the responsibility remains on the user to ensure they aren't causing hinderance to others.
​
You can always use `man` to understand the command or refer to some [cheatsheets](https://www.linuxtrainingacademy.com/linux-commands-cheat-sheet/) for understanding the commands.
​
### Accessing the clusters remotely
After the initial wait time you should be able to log into CCR with your RSA key using `ssh -A UBITNAME@vortex.ccr.buffalo.edu`. Also remember to always be on the VPN or in the UB network else you won't be able to login.
​
### Creating your own anaconda environment with your own dependencies
1. Create a folder path using \
`mkdir -p /projects/academic/courses/cse510/UBITNAME/modulefiles/python` \
and then cd to that directory.
2. Copy lua script by running \
`cp /util/academic/modulefiles/Core/python/anaconda.lua  jupyter.lua`\
It is important to name is jupyter if you choose to use it with the jupyter custom shell in onDemand.
3. Change directory by running `cd /projects/academic/courses/cse510/UBITNAME/`
4. Install anaconda to this folder by running `wget https://repo.anaconda.com/archive/Anaconda3-2020.02-Linux-x86_64.sh`
5. Run chmod to make the shell script runnable and then run it \
`chmod +x Anaconda3-2020.02-Linux-x86_64.sh`\
`./Anaconda3-2020.02-Linux-x86_64.sh`\
You may have to point locations for installing anaconda and stuff. Use the current location by using the `.` to install it in the same folder.
6. After you are done installing you can move on to change your python directory and other things in your lua file. You can do that by running \
`cd /projects/academic/courses/cse510/UBITNAME/modulefiles/python`\
`vi jupyter.lua`\
For people who haven't used vi. Some important commands and how to use can be found [here](https://www.cs.colostate.edu/helpdocs/vi.html)\
Your document needs to look something like this
>whatis([[Anaconda Python 5.2.0 ]])
>setenv("PYTHONDIR","/projects/academic/courses/cse510/UBITNAME")
>prepend_path{"PATH","/projects/academic/courses/cse510/UBITNAME/bin",delim=":",priority="0"}
>prepend_path{"LD_LIBRARY_PATH","/projects/academic/courses/cse510/UBITNAME/lib",delim=":",priority="0"}
>load("cuda/10.1")
>load("ffmpeg/4.0.1")
>if (mode() == "load") then
>  LmodMessage([[ The custom python/anaconda-5.2.0 module by UBITNAME has been loaded. Python 2.7 and 3.6 are available. ]])
>end
​
  You can add more dependent modules like ffmpeg and cuda that I have included by using load(MODULE_NAME). A custom load message also helps identify all installations and configurations are done properly. After this you can just use conda install to load all your requirements to this base installation which will be loaded everytime you use it.
7. Now you have to also change your **.bashrc** script to load your module on startup of your user command line.\
`cd ~`\
`vi .bashrc`\
`module use /projects/academic/courses/cse510/UBITNAME/modulefiles/python`
8. Now you should be able to use this module on every startup of your user account. It does slow down ssh and to initiate a new user session but it works. Just wait for a minute when you ssh.
9. Now when you use onDemand link for which can be found [here](https://ondemand.ccr.buffalo.edu/). You can use the jupyter custom option, on closer inspection of logs you can see your module being loaded. This gives access to all your dependencies like pytorch, tensorflow, etc..
​
### Additional tips to speed up process of writing code and access more storage.
1. When you launch jupyter session it starts your time as soon as the jupyter shell and the process starts. So you may waste time of your session writing the code rather than running it. To prevent this you can actually write the notebook on your laptop with smaller epochs to test everything and then remote copy your code and then run it with higher epochs on the CCR cluster. To remote copy you can take a look at scp an example of running it as below.\
`scp test.ipynb UBITNAME@vortex.ccr.buffalo.edu:~`\
This copies your notebook to your home directory, which is where the jupyter custom shell starts.
2. Since each user is restricted to about 5GB of storage you may not want to use your home directory to store all your data, models and other important things. To solve this you can create a soft link to your shared project space which has higher storage.\
`cd ~`\
`mkdir projectFiles`\
`ln -s /projects/academic/courses/cse510/UBITNAME/projectFiles projectFiles`\
Ensure that the projectFiles is also created in your UBITNAME folder in the shared project space. Now you can just use projectFiles in jupyter and it will actually refer to the shared project space.
