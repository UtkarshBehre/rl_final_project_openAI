## Utility functions

```python
import matplotlib.pyplot as plt

# defining a graph plotting helper method
def plot_graph(xlab, ylab, values, title):
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(title)
    plt.grid(True)
    window_range = 5
    plt.plot([np.mean(values[ind-window_range: ind]) for ind in range(window_range, len(values))])
    plt.show()

def train_and_test_agent(agent_type, env_name, total_episodes, score_limit, test_runs):
    '''
    Initiallizes the agent on env_name, trains it for total_episodes with score_limit, then tests the performance on test_runs
    '''
    if agent_type == "REINFORCE":
        agent = ReinforceAgent(env_name, score_limit)
    elif agent_type == "A2C":
        agent = A2CAgent(env_name, score_limit)
    agent.train_agent(total_episodes)
    agent.test_agent(test_runs)
    plot_graph("Episodes", "Rewards", agent.hist_rewards, "Rewards vs Episodes(agent: %s environment: %s "%(agent.agent_name, agent.env_name))
    print()
    
def train_and_test_agents(env_names, total_episodes, score_limit, test_runs):
    for i in range(len(env_names)):
        train_and_test_agent("REINFORCE", env_names[i], total_episodes, score_limit[i], test_runs)
        train_and_test_agent("A2C", env_names[i], total_episodes, score_limit[i], test_runs)
```
