# Run the space invaders environment and agent files

```python
%run space_invaders_environment.py
%run space_invaders_agent.py
import matplotlib.pyplot as plt

# defining a graph plotting helper method
def plot_graph(xlab, ylab, values, title, takeMean):
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(title)
    plt.grid(True)
    # plots the mean of the next 10 rewards at every episode
    if takeMean:
        window_range = 10
        plt.plot([np.mean(values[ind-window_range: ind]) for ind in range(window_range, len(values))])
    else:
        plt.plot(values)

total_episodes = 10000
epsilon_decay = 0.9997
render_while_training = False
render_while_testing = False
```

# DQN Agent on Space Invaders

```python
env = Environment(env_name="SpaceInvaders-v4")
dqn_agent = Agent_Space_Invaders(env, ddqn=False)
```

```python
dqn_agent.train_agent(total_episodes=total_episodes, epsilon_decay = epsilon_decay, render = render_while_training)
```

```python
dqn_agent.test_agent(render = render_while_testing)
```

```python
plot_graph("episodes", "rewards", dqn_agent.hist_rewards, "Rewards vs Episodes", takeMean=True)
```

```python
plot_graph("episodes", "epsilon", dqn_agent.hist_epsilon_values, "Epsilon Values vs Episodes", takeMean=False)
```
